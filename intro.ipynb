```python
normalized_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(normalized_img)
```

```python
def get_pixel(image, x, y):
    """
    Return the pixel at position (x, y)
    """
    pixel = image[y, x]
    return pixel

def get_image(image, cmap=None):
    """
    Turn of matplot axis
    """
    plt.axis('off')
    plt.imshow(image, cmap=cmap)
    plt.show()
```

```python
get_image(normalized_img)
```

Filters

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
get_image(hsv)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
get_image(gray, cmap='gray')
```

```python
# blur
width, height = normalized_img.shape[:2]
print(width, height)
if(width > 500):
    kernl = (50, 50)
elif (width > 200) and (width <= 500):
    kernl = (25, 25)
else:
    kernl = (10,10)
blur = cv2.blur(normalized_img, kernl)
get_image(blur, cmap='gray')
```

```python
# binary image
v, binary = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
get_image(binary, cmap='gray')
```

```python
# Invert image
invert = cv2.bitwise_not(binary)
get_image(invert, cmap='gray')
```
